```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/weekly_earnings.csv')
```

```python
df.head(5)
```

```python
# drop usesless columns
df.drop(['sid', 'id','position','created_at','created_meta','updated_at','updated_meta','meta','NAICS',], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df_clean = pd.DataFrame()
```

```python
# rename bad province headings
# Define the new column names
new_columns = ['Date', 'Alberta', 'British Columbia', 'Canada', 'Manitoba', 'New Brunswick',
               'Newfoundland and Labrador', 'Nova Scotia', 'Ontario', 'Prince Edward Island',
               'Quebec', 'Saskatchewan']

# Rename the columns in the DataFrame
df.columns = new_columns
```

```python
# Create a list of Canadian provinces by excluding the first column ("When")
provinces = df.columns[1:].tolist()
provinces
```

```python
# Extract year from the "When" column
df['Year'] = pd.to_datetime(df['Date']).dt.year
```

```python
# Initialize an empty DataFrame to store the cleaned data
df_clean = pd.DataFrame(columns=['Date', 'Location', 'Earnings'])

# Iterate over each year and province
for year in df['Year'].unique():

    # Skip the iteration for years beyond 2014
    if year > 2014: #2015 is an incomplete year for an annual calculation
        continue

    for province in provinces:
        # Filter the data for the current year and province
        year_province_data = df[(df['Year'] == year)][['Date', province]]
        
        # Calculate the sum of earnings for the current year and province
        earnings_sum = year_province_data[province].sum() * 4
        
        # Create a temporary DataFrame with the current data
        temp_df = pd.DataFrame({'Date': str(year), 'Location': province, 'Earnings': earnings_sum}, index=[0])
        
        # Concatenate the temporary DataFrame with df_clean
        df_clean = pd.concat([df_clean, temp_df], ignore_index=True)
```

```python
# write dataframe to csv file for Tableau
df_clean.to_csv('/content/drive/MyDrive/Colab Notebooks/datasets/annual_earnings.csv', index=False)
```
